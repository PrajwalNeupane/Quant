```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 100           # Number of time steps
n_paths = 100     # Number of paths to simulate
growth = 1.2
decay = 0.8
threshold = 0.5   # Probability cutoff

# Run simulation
all_paths = []

for _ in range(n_paths):
    y = [1]  # Start at 1
    for i in range(1, n):
        if np.random.random() >= threshold:
            y.append(y[-1] * growth)
        else:
            y.append(y[-1] * decay)
    all_paths.append(y)

# Plotting
for path in all_paths:
    plt.plot(path, alpha=0.3)

plt.title(f"{n_paths} Simulated Multiplicative Paths")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
final_values = [path[-1] for path in all_paths]

mean = np.mean(final_values)
median = np.median(final_values)
std = np.std(final_values)

print(f"Mean final value: {mean:.2f}")
print(f"Median final value: {median:.2f}")
print(f"Standard deviation: {std:.2f}")

# Histogram
plt.hist(final_values, bins=30, edgecolor='black')
plt.title("Distribution of Final Values")
plt.xlabel("Final Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 100               # Time steps
n_paths = 100         # Number of simulations
T = 1.0               # Total time
dt = T / n            # Time step
mu = 0.05            # Drift (e.g., 5% annual return)
sigma = 0.2           # Volatility (e.g., 20% annual volatility)

# Simulate paths
all_paths = []

for _ in range(n_paths):
    S = [1]  # Initial value
    for _ in range(1, n):
        Z = np.random.normal(0, 1)
        S_new = S[-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)
        S.append(S_new)
    all_paths.append(S)

# Plot
for path in all_paths:
    plt.plot(path, alpha=0.3)

plt.title("Simulated GBM Paths")
plt.xlabel("Time Step")
plt.ylabel("Asset Value")
plt.grid(True)
plt.show()
```

```python
log_returns = []

for path in all_paths:
    path = np.array(path)
    returns = np.diff(np.log(path))  # Log returns
    log_returns.extend(returns)

# Histogram of log returns
plt.hist(log_returns, bins=50, edgecolor='black', density=True)
plt.title("Distribution of Log Returns")
plt.xlabel("Log Return")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
